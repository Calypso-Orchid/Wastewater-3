```python
import pandas as pd
treatment = pd.read_csv('wastewater_edited.csv')
treatment = treatment.sort_values(by='% Wastewater Treated', ascending=False)
print(treatment)
```

The above dataframe shows how percentage of wastewater treatment varies amongst countries. 
El Salvador has the lowest percentage of population connected at 2.83 percent.
Singapore and Kuwait have 100 percent of their population connected to wastewater treated.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = treatment['Country or Area']
x1 = treatment['Continent']
y = treatment['% Wastewater Treated']
pop = treatment['Pop (2012)']
```

```python
plt.plot(y,pop)
plt.xlabel('% Wastewater Treated')
plt.ylabel('Population 2012')
plt.title('Figure 1: Population vs. % Wastewater Treated')
plt.show()
```

Figure 1. In this data, the country with the highest population has only about half of its wastewater treated. 

```python
colors = treatment['% Wastewater Treated']
area = np.pi * (treatment['% Wastewater Treated'])
plt.scatter(x1, y, s=area, c=colors, alpha=0.5)
plt.xlabel('Continent')
plt.ylabel('% Wastewater Treated')
plt.title('Figure 2: Continent vs. % Wastewater Treated')
plt.show()
```

Figure 2. Wastewater treatment varies among countries within continents. 

```python
treatment.groupby('Continent').mean()
```

```python
x3 = (4.90, 83.34, 19.51, 63.01, 82.00)
y3 = ('Africa', 'Asia', 'Central America', 'Europe', 'Oceania')
plt.bar(y3,x3) 
plt.xlabel('Continent')
plt.ylabel('Mean % of Wastewater Treated')
plt.title('Figure 3: Continent vs. Mean % of Wastewater Treated')
plt.show()
```

Figure 3. Africa and Central America have the lowest means in wastewater treated. Asia has the highest mean percentage of wastewater treated.
